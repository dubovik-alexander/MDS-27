### Задание 1


Посчитайте значение первой и второй производной функции:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$ 

В точке x = 1.


```python
from sympy import diff, Symbol, cos, sin, evalf

```

### Решение

#### Вариант 1

```python
x = Symbol('x')
```

```python
def fx(x):
    return diff(x**5 + 4*sin(2*x)+cos(3*x+3))
```

```python
fx(x).evalf(subs = {'x':1})
```

```python
def fxx(x):
    return diff(fx(x))
```

```python
fxx(x).evalf(subs = {'x':1})
```

#### Вариант 2

```python
def fx2(x):
    return x**5 + 4*sin(2*x)+cos(3*x+3)

def dfx(x, z = fx2):
    h = 1e-5
    return (z(x+h)-z(x-h))/(2*h)

print( f'Значение первой производной: {dfx(1)}, Значение второй производной {dfx(1, dfx)}')
```

#### Вариант 2

```python
from math import *
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (sin(2*x+1))**5

def dfx(x, z = f):
    h = 1e-5
    return (z(x+h)-z(x-h))/(2*h)


x = np.linspace(-5, 5, 100)

x1 = []
y1 = []
for i in range(100):
    dfx(x[i])
    # print('x = ',x[i],', y = ',f(x[i]))
    x1.append(x[i])
    y1.append(dfx(x[i]))

    
x2 = []
y2 = []
for i in range(100):
    dfx(x[i], dfx)
    # print('x = ',x[i],', y = ',f(x[i]))
    x2.append(x[i])
    y2.append(dfx(x[i], dfx))
```

```python
plt.plot(x1, y1)
plt.plot(x2, y2)
plt.grid(True)
plt.show()
```
