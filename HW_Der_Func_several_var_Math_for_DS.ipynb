## –£—Ä–æ–≤–µ–Ω—å 0:


### –ó–∞–¥–∞–Ω–∏–µ 1

–ü–æ—Å—á–∏—Ç–∞–π—Ç–µ —á–∞—Å—Ç–Ω—ã–µ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—ã–µ —Ñ—É–Ω–∫—Ü–∏–π:

1) $f(x,y)=2x^2y^3 + 1/x + y^2x + 7$

2) $f(x,y)=x^2y - sin(xy) + cos(x^2) + 6y$

___

```python
import sympy as sym
```

```python
x = sym.Symbol('x')
y = sym.Symbol('y')
```

#### <span style="color:blue"> 1) $f(x,y)=2x^2y^3 + 1/x + y^2x + 7$</span>

```python
fxy1 = 2*(x**2)*(y**3) + 1/x + (y**2)*x + 7
```

#### –ù–∞—Ö–æ–¥–∏–º —á–∞—Å—Ç–Ω—ã–µ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—ã–µ:

```python
sym.diff(fxy1, x) # –¥–∏—Ñ–∏—Ä–∏–Ω—Ü–∏—Ä—É–µ–º –ø–æ —Ö
```

```python
sym.diff(fxy1, y) # –¥–∏—Ñ–∏—Ä–∏–Ω—Ü–∏—Ä—É–µ–º –ø–æ y
```

#### –ù–∞–π–¥–µ–º —Å–º–µ—à–∞–Ω–Ω—ã–µ —á–∞—Å—Ç–Ω—ã–µ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—ã–µ (–Ω–µ –ø–æ –∑–∞–¥–∞–Ω–∏—é, –∞ –¥–ª—è –ø–æ–Ω–∏–º–∞–Ω–∏—è –∫–∞–∫ —ç—Ç–æ —Ä–∞–±–æ—Ç–∞–µ—Ç):

```python
sym.diff(sym.diff(fxy1, x), y) # –¥–∏—Ñ–∏—Ä–∏–Ω—Ü–∏—Ä—É–µ–º —Å–Ω–∞—á–∞–ª–∞ –ø–æ —Ö, –ø–æ—Ç–æ–º –ø–æ y
```

___

#### <span style="color:blue"> 2) $f(x,y)=x^2y - sin(xy) + cos(x^2) + 6y$ </span>

```python
fxy2 = (ùë•**2)*ùë¶ - sym.ùë†ùëñùëõ(ùë•*ùë¶) + sym.ùëêùëúùë†(ùë•**2) + 6*ùë¶
```

#### –ù–∞—Ö–æ–¥–∏–º —á–∞—Å—Ç–Ω—ã–µ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—ã–µ:

```python
sym.diff(fxy2, x) # –¥–∏—Ñ–∏—Ä–∏–Ω—Ü–∏—Ä—É–µ–º –ø–æ —Ö
```

```python
sym.diff(fxy2, y) # –¥–∏—Ñ–∏—Ä–∏–Ω—Ü–∏—Ä—É–µ–º –ø–æ y
```

#### –ù–∞–π–¥–µ–º —Å–º–µ—à–∞–Ω–Ω—ã–µ —á–∞—Å—Ç–Ω—ã–µ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—ã–µ (–Ω–µ –ø–æ –∑–∞–¥–∞–Ω–∏—é, –∞ –¥–ª—è –ø–æ–Ω–∏–º–∞–Ω–∏—è –∫–∞–∫ —ç—Ç–æ —Ä–∞–±–æ—Ç–∞–µ—Ç):

```python
sym.diff(sym.diff(fxy2, x), y) # –¥–∏—Ñ–∏—Ä–∏–Ω—Ü–∏—Ä—É–µ–º —Å–Ω–∞—á–∞–ª–∞ –ø–æ —Ö, –ø–æ—Ç–æ–º –ø–æ y
```

___

## –£—Ä–æ–≤–µ–Ω—å 1:

### –ó–∞–¥–∞–Ω–∏–µ 2

–í—ã—á–∏—Å–ª–∏—Ç–µ –≥—Ä–∞–¥–∏–µ–Ω—Ç —Ñ—É–Ω–∫—Ü–∏–∏:

$f(x,y)=2x^2y^3 + 1/x + y^2x + 7$

–≤ —Ç–æ—á–∫–µ $(1,2)$

```python
i = sym.Symbol('i')
j = sym.Symbol('j')
```

```python
fxy3 = 2*(ùë•**2)*(ùë¶**3) + 1/ùë• + (ùë¶**2)*ùë• + 7
```

```python
# –ù–∞—Ö–æ–¥–∏–º —á–∞—Å—Ç–Ω—ã–µ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—ã–µ:
```

```python
sym.diff(fxy3, x) # –¥–∏—Ñ–∏—Ä–∏–Ω—Ü–∏—Ä—É–µ–º –ø–æ —Ö
```

```python
sym.diff(fxy3, y) # –¥–∏—Ñ–∏—Ä–∏–Ω—Ü–∏—Ä—É–µ–º –ø–æ y
```

```python
# –¢–æ–≥–¥–∞ –≤–µ–ª–∏—á–∏–Ω–∞ –≥—Ä–∞–¥–∏–µ–Ω—Ç–∞ —Ä–∞–≤–Ω–∞:
    
grad_fxy3 = sym.diff(fxy3, x) * i + sym.diff(fxy3, y) * j
grad_fxy3
```

```python
grad_fxy3.evalf(subs = {'x':1, 'y':2})
```

### –ó–∞–¥–∞–Ω–∏–µ 3

–ü–æ—Å—Ç—Ä–æ–π—Ç–µ —Ç—Ä–µ—Ö–º–µ—Ä–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏

$f(x)=x_2^2-x_1^2 x_2 + 2 x_1 x_2$

```python
### –î–ª—è —É–ø—Ä–æ—â–µ–Ω–∏—è —Ö1 = —Ö, —Ö2 = y
```

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from matplotlib import cm
import matplotlib.pyplot as plt
```

```python
# —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏
f = lambda x, y: y**2 - (x**2)*y + 2*x*y
```

```python

fig = plt.figure(figsize = (10, 10))
ax = fig.add_subplot(1, 1, 1, projection = '3d')

# —Ä–∞–∑–º–µ—á–∞–µ–º –≥—Ä–∞–Ω–∏—Ü—ã –æ—Å–µ–π –¥–ª—è –∞—Ä–≥—É–º–µ–Ω—Ç–æ–≤
xval = np.linspace(-4, 6, 100)
yval = np.linspace(-5, 5, 100)


x, y = np.meshgrid(xval, yval)
# –ø—Ä–∏—Ä–∞–≤–Ω–∏–≤–∞–µ–º z –∫ —Ñ—É–Ω–∫—Ü–∏–∏ –æ—Ç x –∏ y 
z = f(x, y)

# –ì—Ä–∞—Ñ–∏–∫ –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏
surf = ax.plot_surface(x, y, z, rstride = 5, cstride = 5, cmap = cm.plasma)

# –¶–≤–µ—Ç–æ–≤–∞—è –ª–µ–≥–µ–Ω–¥–∞
fig.colorbar(surf, shrink=0.5, aspect=5)
```
